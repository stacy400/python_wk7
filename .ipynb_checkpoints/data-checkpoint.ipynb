```python
cols = ["DATE_DIED"] + [col for col in df.columns if col != "DATE_DIED"]
df = df[cols]
df
```

```python
# Line Chart: Deaths over Time
import matplotlib.pyplot as plt
deaths_over_time = df["DATE_DIED"].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.plot(deaths_over_time.index, deaths_over_time.values, color="red", marker="o", label="Deaths")
plt.title("COVID-19 Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Deaths")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Bar Chart: Average Age by Patient Type
avg_age = df.groupby("PATIENT_TYPE")["AGE"].mean()
plt.figure(figsize=(7,5))
avg_age.plot(kind="bar", color="blue")
plt.title("Average Age by Patient Type")
plt.xlabel("Patient Type")
plt.ylabel("Average Age")
plt.xticks(rotation=0)
plt.show()
```

```python
#Histogram: Age Distribution
plt.figure(figsize=(8,5))
plt.hist(df["AGE"].dropna(), bins=20, color="purple", edgecolor="black")
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis="y")
plt.show()
```

```python
#Scatter Plot: Age vs ICU Admission
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(df["AGE"], df["ICU"], color="green", alpha=0.5)
plt.title("Age vs ICU Admission")
plt.xlabel("Age")
plt.ylabel("ICU Status (Admitted / Not admitted)")
plt.grid(True)
plt.show()
```
